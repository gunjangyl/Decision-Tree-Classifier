```python
#Import Essential Libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
#Load Iris Dataset
iris=load_iris()
 #(from sklearn library)
```

```python
#Extracting independent & dependent variables
X=iris.data # Independent variable
y=iris.target # Dependent variable
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#Plotting Decision Tree
from sklearn.tree import plot_tree
plt.figure(figsize=(12,8)) #12rows & 8columns(grid system)
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)  #clf is the model of DecisionTreeClassifier(iris dataset is used)
#feature_names are: sepal-length, sepal-width, petal-length, petal-width
#class_names are:  Setosa, Virginia, Versicolor
#filled: used for filling by-default colors
plt.title("Decision Tree")
plt.show()
```
